$$
\sqrt(x ^ 2 + y ^2)\\
||\mathbf{x}|| = \sqrt{\sum_i x_i ^ 2}
$$

```python
def norm(x):
    acum = 0
    for x_i in x:
        acum += x_i ** 2
    return acum ** 0.5
```

- generar un arreglo uniform de size (1000, 10)
- Todas las columnas impares se vuelvan negativas.
- Sumar todo el arreglo

1. for loop

```python
%%timeit
x = np.random.uniform(size=(1000, 10))
for i in range(x.shape[1]):
    if i % 2:
        x[:, i] *= -1
acum = np.sum(x)
```

2. Indexacion

```python
%%timeit
x = np.random.uniform(size=(1000, 10))
x[:, 1::2] *= -1
acum = x.sum()
```
